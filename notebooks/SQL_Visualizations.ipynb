### Cohort Analysis

```python
cohort_analysis=pd.read_csv(r"C:\Users\Pavitra\Downloads\olist-Project\exports\cohort_analysis.csv")
cohort_analysis.head()
```

```python


cohort_analysis['cohort_month'] = pd.to_datetime(
    cohort_analysis['cohort_month']
)

cohort_analysis = cohort_analysis.sort_values('cohort_month')

months = cohort_analysis['cohort_month'].dt.strftime('%b-%Y')

x = np.arange(len(months))
width = 0.4

fig, ax1 = plt.subplots(figsize=(15,10))

# Total Revenue (Left Axis)
bars1 = ax1.bar(
    x - width/2,
    cohort_analysis['revenue_per_cohort'],
    width,
    label='Total Revenue per Cohort',
    color='steelblue'
)

ax1.set_xlabel('Cohort Month')
ax1.set_ylabel('Total Revenue per Cohort')
ax1.set_xticks(x)
ax1.set_xticklabels(months, rotation=45)

# Average Revenue (Right Axis)
ax2 = ax1.twinx()

bars2 = ax2.bar(
    x + width/2,
    cohort_analysis['average_revenue_per_customer'],
    width,
    label='Average Revenue per Customer',
    color='orange'
)

ax2.set_ylabel('Average Revenue per Customer')

# Combined Legend
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(
    handles1 + handles2,
    labels1 + labels2,
    bbox_to_anchor=(1.02, 1),
    loc='upper left'
)

plt.title('Revenue Metrics by Cohort Month')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.bar(
    cohort_analysis["cohort_month"],
    cohort_analysis["number_of_customer"],
    color='steelblue'
)

plt.xlabel("Cohort Month")
plt.ylabel("Number of Customers")
plt.title("Number of Customers by Cohort Month")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
### Key Findings

* The business grew rapidly from **early 2017 to late 2017**, with both the number of customers and total revenue increasing steadily month after month.

* **November 2017** was the strongest cohort, bringing in the highest revenue and one of the largest customer groups. This was likely influenced by major shopping events or successful marketing campaigns.

* Although total revenue increased a lot, the **average revenue per customer stayed fairly stable** across most cohorts. This means customers were spending about the same amount on average, and growth mainly came from attracting more customers.

* During **2018**, customer acquisition remained strong, but growth started to level off compared to the rapid increase seen in 2017.

### Business Suggestions

* Focus on **increasing the amount each customer spends**, since customer spending has remained relatively flat. Product bundles, recommendations, and discounts on larger purchases could help.

* Study what worked during the **late 2017 high-growth period** and try to repeat similar campaigns or promotions.

* Invest more in **customer retention and repeat purchases**, as bringing existing customers back is often cheaper than constantly acquiring new ones.

* Identify and target the most valuable customer segments with personalized offers to improve overall revenue.
```

```python
retention_df = pd.read_csv(r"C:\Users\Pavitra\Downloads\olist-Project\exports\cohort_retention_analysis.csv")
retention_df
```

```python
# Create pivot table
retention_matrix = retention_df.pivot(
    index='cohort_month',
    columns='months_since_first_purchase',
    values='retention_rate'
)

# Remove Month 0 (100% retention)
retention_matrix = retention_matrix.drop(columns=[0])

plt.figure(figsize=(12,8))

sns.heatmap(
    retention_matrix,
    annot=True,
    fmt=".2f",
    cmap="Blues",
    vmin=0,
    vmax=1      # adjust based on your max retention
)

plt.title("Cohort Retention Heatmap (%)")
plt.xlabel("Months Since First Purchase")
plt.ylabel("Cohort Month")

plt.tight_layout()
plt.savefig("cohort_retention_heatmap.png")
plt.show()
```

** What the heatmap is saying

Month 0 is always 100% because that is the first purchase month of each cohort.
After that, retention drops very fast. Most cohorts fall to very low values by month 1 and stay low.
Some cohorts in late 2017 / early 2018 look a bit stronger in the first few months, but the improvement is small.
The last cohorts look shorter because there are not enough future months left in the dataset.

Business problems seen here

Very low repeat buying. Most customers buy once and do not come back much.
Revenue depends more on new customers than on loyal repeat customers.
Later cohorts are incomplete, so long-term retention for 2018 cannot be judged fully.
Customer engagement after first purchase looks weak, which usually means weak follow-up or poor return incentive.

Business conclusions

The business is good at getting first-time buyers, but not good at bringing them back.
Retention is the main weak point, not customer acquisition.
A few cohorts perform a little better early on, so there may be some months or campaigns that worked better than others.

Suggestions to improve

Send follow-up offers after the first order like coupons or free shipping on the next purchase.
Use reminder emails / WhatsApp / app messages within 7–30 days after purchase.
Study the better cohorts from late 2017 and copy what worked there.
Improve repeat purchase reasons: bundles, loyalty points, personalized recommendations, and fast support.

Short takeaway

The business is bringing people in, but keeping them is the big problem.
Fixing retention should give a much bigger lift than only chasing new customers.

** Key Findings
 
25%of orders were delivered within 7 days.
50% of orders were delivered within 10 days (median delivery time).
75% of orders were delivered within 16 days.
90% of orders were delivered within 23 days.

Business Interpretation

Most customers received their orders in about 1–2 weeks, which is fairly reasonable for a large e-commerce platform.
The median (10 days) is lower than the average (12.5 days), which suggests a small number of very late deliveries are pulling the average up.
The jump from 16 days (75th percentile) to 23 days (90th percentile) shows that a small group of customers experience noticeably slower deliveries.

Problems Identified

Around 10% of customers wait more than 23 days, which can negatively affect customer satisfaction and reviews.
Extremely delayed orders (such as the 210-day outlier found earlier) indicate operational exceptions that need attention.
Delivery experience is not fully consistent across all customers.

Suggestions for Improvement

Focus on the slowest 10% of deliveries, since they contribute most to customer frustration.
Identify regions, products, or logistics partners associated with long delivery times.
Set internal delivery targets such as delivering 90% of orders within 20 days.
Provide proactive updates and tracking information when delays occur.

5-star reviews dominate the dataset, making up the majority of all reviews.

Most customers are satisfied with their purchases, products, and delivery experience.
The high number of 4-star and 5-star reviews indicates a healthy customer experience.
However, the sizeable group of 1-star reviews should not be ignored because these customers are the most likely to churn and leave negative feedback.

Potential Problems

Some customers are having very poor experiences, leading directly to 1-star reviews.
These issues may be related to delayed deliveries, damaged products, incorrect items, or customer service problems.
Negative reviews can affect brand reputation and future sales.

Suggestions

Analyze 1-star reviews separately to identify the most common complaints.
Contact dissatisfied customers quickly and offer support, refunds, or replacements when appropriate.
Study what drives 5-star reviews and try to replicate those factors across more orders.
Monitor delivery delays and product quality, since these are common causes of poor ratings.

** Key Findings
 
Most customers gave high ratings (4–5 stars), showing generally good customer satisfaction.
However, repeat purchase rates are low across all groups (around 2–5%).
Customers with medium ratings (3 stars) have the highest repeat purchase rate (5.4%), which is unexpected.
Customers with high ratings buy slightly more often than low-rating customers, but the difference is small.

Business Interpretation

Customer satisfaction alone does not seem to be a strong driver of repeat purchases in this dataset.
Many customers appear to buy only once, regardless of whether they were satisfied or dissatisfied.
This suggests that factors like product type, purchase frequency, promotions, or customer needs may have a bigger impact on repeat buying than review scores.

Possible Business Problems

The business may have a one-time purchase pattern, where customers buy only when they need something.
There may be limited customer retention efforts after the first purchase.
High satisfaction is not being converted into customer loyalty.

Suggestions

Target happy customers (4–5 stars) with loyalty programs, coupons, and personalized recommendations to encourage another purchase.
Follow up with 3-star reviewers, since they surprisingly show the highest repeat rate and may be easier to convert into loyal customers.
Investigate why repeat purchase rates are low overall, even among satisfied customers.
Combine review scores with other factors (delivery speed, order value, product category) to better understand customer retention
